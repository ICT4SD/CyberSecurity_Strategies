```python
# import packages
import sys

import json

import re

import string

import nltk

import uuid
```

```python
try:

    path = sys.argv[1]

    passage = open("C:\\Users\\Shimeng Lin\\Desktop\\Fordham\\Spring 2017\\United Nations Cybersecurity Research Project\\raw data_UK_txt\\UK.txt",'r').read()

except:

    print ('Not Valid File Path.\n','Use', sys.argv[0],'<C:\\Users\\Shimeng Lin\\Desktop\\Fordham\\Spring 2017\\United Nations Cybersecurity Research Project\\raw data_UK_txt\\UK.txt>')

    sys.exit(2)

passage =''.join(filter(lambda x: x in string.printable, passage.replace('\n',' ')))
```

```python
content_pattern = re.compile("[0-9]+\.$")

# create a list called sentences
sentences = list()

for sen in nltk.sent_tokenize(passage): 

    sen = ' '.join(sen.split())

    if not content_pattern.search(sen) and len(sen) > 25: 

        sentences.append(sen)      
```

```python
sentence_id = [(uuid.uuid4().hex[:10],i) for i in sentences] 
```
